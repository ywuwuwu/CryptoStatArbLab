# 02 Results Review
Load sweep results and visualize key tradeoffs.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_parquet('runs/latest/results.parquet')
df.sort_values('net_sharpe', ascending=False).head(10)
```

```python
plt.figure()
plt.scatter(df['avg_turnover'], df['net_sharpe'], s=12)
plt.xlabel('avg_turnover')
plt.ylabel('net_sharpe')
plt.title('Turnover vs Net Sharpe')
plt.show()
```
